```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca file CSV
file_path = 'C:/Users/ASUS/Documents/Afwan/capstone_project/data/Bungkuk/Bungkuk Wan 27 - Sheet1.csv'  # Ganti dengan path file kamu jika berbeda
df = pd.read_csv(file_path)

# Visualisasi data akselerometer
plt.figure(figsize=(12, 6))
plt.plot(df['Accel_X'], label='Accel_X')
plt.plot(df['Accel_Y'], label='Accel_Y')
plt.plot(df['Accel_Z'], label='Accel_Z')
plt.plot(df['Gyro_X'], label='Gyro_X')
plt.plot(df['Gyro_Y'], label='Gyro_Y')
plt.plot(df['Gyro_Z'], label='Gyro_Z')

# Menambahkan judul dan label
plt.title('Visualisasi Data IMU')
plt.xlabel('Index Data')
plt.ylabel('Nilai Akselerasi')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Tampilkan grafik
plt.show()
```
